```python
import tensorflow as tf
import tensorflow.keras.layers as layers
import tensorflow.keras.models as models
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
from sklearn.model_selection import train_test_split
```

```python
train_labels = pd.read_csv('train_labels.csv',dtype= {'Id': 'str', 'Category': 'str'})
```

```python
IMAGE_SIZE = 224
```

```python
X = []
y = []
for i in range(train_labels.shape[0]):
    img = cv2.imread('train/{}.jpg'.format(train_labels.iloc[i]['Id']))
    X.append(cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE)))
    y.append(train_labels.iloc[i]['Category'])
```

```python
from tensorflow.keras.utils import to_categorical
classes = {'rose': 0, 'sunflower': 1, 'daisy': 2, 'tulip': 3, 'dandelion': 4}
yi = [classes[s] for s in y]
y_sparse = to_categorical(yi)
```

```python
X = np.stack(X).astype(np.float32)/255.0
train_X, val_X, train_Y, val_Y = train_test_split(X, y_sparse, test_size=0.1, random_state=17, stratify=y_sparse)
```

```python
base_model = tf.keras.applications.InceptionV3(weights='imagenet', include_top=False)

x = base_model.output
x = layers.Dropout(0.2)(x)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(512, activation='relu')(x)
x = layers.Dropout(0.1)(x)
predictions = layers.Dense(len(classes), activation='softmax')(x)

m = models.Model(inputs=base_model.input, outputs=predictions)

for layer in base_model.layers:
    layer.trainable = False
```

```python
m.compile(optimizer='adam', 
          loss='categorical_crossentropy',
          metrics=['accuracy'])
```

```python
m.fit(train_X, train_Y, 
      validation_data=(val_X, val_Y),
      batch_size=8,
      epochs = 200000,
      callbacks=[
          tf.keras.callbacks.TensorBoard(log_dir='./Graph-2', histogram_freq=0, write_graph=True, write_images=True),
          tf.keras.callbacks.ModelCheckpoint("saved-model-{epoch:05d}-{val_acc:.2f}.hdf5", monitor='val_loss', verbose=0, save_best_only=False, save_weights_only=False, mode='auto', period=100)
      ])
```
